## ch2.3 手写数字识别

```python
from os import listdir
import numpy as np
from knn import classify0

# 允许一个 cell 有多个输出
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

### 准备数据

```python
def img2vector(filename):
    vect = np.zeros((1, 1024))
    with open(filename) as f:
        for i in range(32):
            line = f.readline()
            for j in range(32):
                vect[0, 32*i+j] = int(line[j])
    return vect

test_vect = img2vector('../^src/Ch02/digits/testDigits/0_13.txt')
test_vect[0, 0:31]
test_vect[0, 32:63]
```

### 测试分类器

```python
training_dir = '../^src/Ch02/digits/trainingDigits'
test_dir = '../^src/Ch02/digits/testDigits'

def handwriting_test():
    labels = []
    training_files = listdir(training_dir)
    m = len(training_files)
    training_mat = np.zeros((m, 1024))
    for i in range(m):
        filename = training_files[i]
        file_basename = filename.split('.')[0]
        class_num = int(file_basename.split('_')[0])
        labels.append(class_num)
        training_mat[i, :] = img2vector('{}/{}'.format(training_dir, filename))
    
    test_files = listdir(test_dir)
    error_count = 0
    m_test = len(test_files)
    for i in range(m_test):
        filename = test_files[i]
        file_basename = filename.split('.')[0]
        class_num = int(file_basename.split('_')[0])
        vector_under_test = img2vector('{}/{}'.format(test_dir, filename))
        classifier_result = classify0(vector_under_test, training_mat, labels, 3)
        print('predict: {}, true: {}'.format(classifier_result, class_num))
        if (classifier_result != class_num):
            error_count += 1
    print('\nthe total number of errors is: {}'.format(error_count))
    print('the total error rate is: {}'.format(error_count/m_test))

handwriting_test()
```

### note
kNN 算法计算效率比较低, 因为对每个测试样本都需要与所有训练样本计算距离. 一种改进的方法是利用 kd-tree 数据结构.
